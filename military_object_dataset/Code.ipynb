```python
import os
from ultralytics import YOLO
import cv2

dataset_dir = r"C:\Military_object_Detection\military_object_dataset"
# ------------------------

yaml_path = os.path.join(dataset_dir, "military_dataset.yaml")

# Read the original YAML
with open(yaml_path, "r") as f:
    lines = f.readlines()

# Update the paths to point to your local absolute directories
with open(yaml_path, "w") as f:
    for line in lines:
        if line.startswith("train:"):
            # Using os.path.join ensures the slashes are correct for your OS
            f.write(f"train: {os.path.join(dataset_dir, 'train', 'images')}\n")
        elif line.startswith("val:"):
            f.write(f"val: {os.path.join(dataset_dir, 'val', 'images')}\n")
        elif line.startswith("test:"):
            f.write(f"test: {os.path.join(dataset_dir, 'test', 'images')}\n")
        else:
            f.write(line)

print(f"Updated YAML saved successfully at: {yaml_path}")
```

```python
import tkinter as tk
from tkinter import filedialog
from ultralytics import YOLO

# ── Pick file via dialog ──────────────────────────────────────
root = tk.Tk()
root.withdraw()

SOURCE = filedialog.askopenfilename(
    title="Select Image or Video",
    filetypes=[
        ("All supported", "*.jpg *.jpeg *.png *.bmp *.mp4 *.avi *.mov *.mkv"),
        ("Images",        "*.jpg *.jpeg *.png *.bmp"),
        ("Videos",        "*.mp4 *.avi *.mov *.mkv"),
    ]
)

if not SOURCE:
    print("No file selected.")
else:
    print(f"Selected: {SOURCE}")
    MODEL_PATH = r"C:\Military_object_Detection\military_object_dataset\runs\detect\YOLO_FYP\train_backup\weights\best.pt"
    model = YOLO(MODEL_PATH)
    model.predict(
        source   = SOURCE,
        conf     = 0.4,
        imgsz    = 640,
        save     = True,
        show     = True,
        device   = 0,
        project  = r"C:\Military_object_Detection\inference_results",
        name     = "run",
        exist_ok = True,
    )
    print(f"Done! Results saved to: C:\\Military_object_Detection\\inference_results\\run")
```
